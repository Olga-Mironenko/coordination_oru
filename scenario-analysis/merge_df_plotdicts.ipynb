```python
import glob
import pandas as pd

dir = 'images/20241203_170129_all600'

# Step 1: Define a list of file paths
file_paths = sorted(glob.glob(f'{dir}/df_plotdicts_map*.csv'))

# Step 2: Read each file and reconstruct the MultiIndex DataFrame
dfs = []
for file in file_paths:
    # Read the CSV
    df = pd.read_csv(file, header=[0, 1], index_col=[0, 1, 2])  # Adjust header and index_col to match the file
    dfs.append(df)

# Step 3: Merge all DataFrames
merged_df = pd.concat(dfs)

# Ensure the merged DataFrame has sorted indices for consistency
merged_df.sort_index(inplace=True)

# Step 4: Save the merged DataFrame if needed
merged_df.to_csv(f'{dir}/df_plotdicts_all.csv')

# Print or use the merged DataFrame
merged_df
```
